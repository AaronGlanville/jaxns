# Bayesian Mixture of Experts

We'll solve a classification problem using mixture of experts. An expert shall be parametrised function that computes a class probability distribution. A mixture of experts is said to be a convex sum of expert outputs.

Let an expert be $f_k: \mathcal{X} \to \mathcal{S}_n$ be $k$-th expert, where $\mathcal{X}$ is a set of objects that can be classified, and $\mathcal{S}_n$ is the unit-simplex over $n$ classes.

We define a distribution over $m$ experts with a Multinomial random variable, $k \mid w \sim P(k \mid w) = \mathrm{Mult}(k \mid w)$, with expert prior probability Dirichlet random variable, $w \in \mathcal{S}_m$, where $w \mid \alpha \sim P(w \mid \alpha) = \mathrm{Dir}(w \mid \alpha)$. We'll place a prior on the concentration parameter as $\log \alpha_k \sim P(\alpha_k) = \mathcal{N}(0, 1)$ for $k=1..m$.

The likelihood of the data, given a particular expert $k$ is given by a product of multinomial distributions, $\mathcal{L}(d_i \mid k)$. Note, for a single observation this reduces to $\mathcal{L}(d_i \mid k) = f^{d_i}_k(x_i)$. Where the superscript represents the component extraction.

$$\mathcal{L}(\mathcal{D} \mid k) = \prod_{i=1}^{|\mathcal{D}|} \mathcal{L}(d_i \mid k)\\
= \prod_{i=1}^{|\mathcal{D}|} f^{d_i}_k(x_i)$$

We will choose $f_k$ so as to maximise the Bayesian evidence,

$$Z = \int \mathcal{L}(\mathcal{D} \mid k) dP(k)\\
= \int \prod_{i=1}^{|\mathcal{D}|} f^{d_i}_k(x_i) dP(k)\\
= \sum_k \int \int \prod_{i=1}^{|\mathcal{D}|} f^{d_i}_k(x_i) P(k | w) P(w | \alpha) P(\alpha)\,\mathrm{d}\alpha \mathrm{d}w\\
= \sum_k \int \int \prod_{i=1}^{|\mathcal{D}|} f^{d_i}_k(x_i) \mathrm{Multi}(k \mid w) \mathrm{Dir}(w \mid \alpha) \mathcal{N}(\log \alpha_k \mid 0, 1)\,\mathrm{d}\alpha \mathrm{d}w$$

We will iteratively determine $Z$ then maximise $Z$ over $f_k$, using the representation $Z = \sum_i L_i dX_i$. That is, we'll hold $dX_i$ fixed and maximise the sum over the parameters of $f_k$.

For the experts, $f_k$, we'll use neural networks.

```python
# Generate data

def xor_reduce(x):
    """
    Computes the XOR reduction on a sequence of bits.

    Examples:
        100 -> xor(xor(1,0),0) = 1
        001 -> xor(xor(0,0),1) = 1
        110 -> xor(xor(1,1),0) = 0
        011 -> xor(xor(0,1),1) = 0

    Args:
        x: boolean vector of bits.

    Returns:
        bool, scalar
    """
    output = x[0]
    for i in range(1, x.shape[-1]):
        output = jnp.logical_xor(output, x[i])
    return output


num_variables = 7
options = [True, False]
x = jnp.asarray(list(product(options, repeat=num_variables)))  #N,2
y = vmap(xor_reduce)(x)  #N
x = x.astype(jnp.float32)
print("Data:")

for input, output in zip(x, y):
    print(f"{input} -> {output}")
```

```python
from jaxns import UnnormalisedDirichlet, Categorical
from jaxns import Prior, Model
from jax.scipy.special import logsumexp
num_experts = 2
num_classes = 2


# Define the likelihood, using Haiku as our framework for neural networks

def expert(x):
    mlp = hk.Sequential([hk.Linear(4),
                         jax.nn.sigmoid,
                         hk.Linear(num_classes)])
    return mlp(x)


def mixture_of_experts(x, k):
    y = []
    for _ in range(num_experts):
        y.append(expert(x))
    y = jnp.stack(y, axis=0)
    return y[k]


mixture_of_experts = hk.without_apply_rng(hk.transform(mixture_of_experts))
# We must call the model once to get the params shape and type as a big pytree
# We then use ravel_pytree to flatten and get the unflatten function.
init_params = mixture_of_experts.init(random.PRNGKey(2345), x, jnp.asarray(0))
init_params_flat, unravel_func = ravel_pytree(init_params)
n_dims = init_params_flat.size
print("Number of parameters:", n_dims)

def build_model(params):
    def log_likelihood(k):
        params_dict = unravel_func(params)
        logits = mixture_of_experts.apply(params_dict, x, k) # N, 2
        logits -= logsumexp(logits, axis=-1, keepdims=True)
        log_prob_true = logits[:, y.astype(jnp.int32)] # N
        return jnp.sum(log_prob_true)

    def prior_model():
        log_alpha = yield Prior(tfpd.Normal(jnp.zeros(num_classes), 1. * jnp.ones(num_classes)), name='log_alpha')
        alpha = jnp.exp(log_alpha)
        w = yield UnnormalisedDirichlet(concentration=alpha, name='w')
        w = w/jnp.sum(w)
        k = yield Categorical(parametrisation='gumbel_max', probs=w, name='k')
        return k


    model = Model(prior_model=prior_model, log_likelihood=log_likelihood)
    return model
```

```python
from jax import value_and_grad
from jax._src.ops.special import logsumexp

num_bayesian = 5
perm = random.permutation(random.PRNGKey(42), jnp.arange(init_params_flat.size))
basyesian_dims, map_dims = perm[:num_bayesian], perm[num_bayesian:]


def mod_log_likelihood(bayes_params, map_params):
    """
    log(P(y|p))
    p = exp(logits)/1 - exp(logits)
    = log(p) * y + log(1-p) * (1-y)
    = logits * y1 - log(exp(-logits)/(exp(-logits) - 1)) * y0
    """
    params = init_params_flat.at[basyesian_dims].set(bayes_params)
    params = params.at[map_dims].set(map_params)
    return log_likelihood(params)


def evidence_maximisation_iter(params):
    map_params = params[map_dims]

    def prior_model():
        bayes_params = yield Prior(tfpd.Uniform(-10 * jnp.ones(num_bayesian), 10. * jnp.ones(num_bayesian)),
                                   name='params')
        return bayes_params, map_params

    model = Model(prior_model=prior_model, log_likelihood=mod_log_likelihood)

    model.sanity_check(random.PRNGKey(0), S=100)

    # Create the nested sampler class. In this case without any tuning.
    exact_ns = ExactNestedSampler(model=model, num_live_points=model.U_ndims * 20, max_samples=1e6,
                                  num_parallel_samplers=len(jax.devices()))

    termination_reason, state = exact_ns(random.PRNGKey(42), term_cond=TerminationCondition(live_evidence_frac=1e-5))
    results = exact_ns.to_results(state, termination_reason)
    exact_ns.summary(results)

    samples = resample(random.PRNGKey(0), results.samples, log_weights=results.log_dp_mean, S=100)

    def loss(map_params):
        return -logsumexp(vmap(lambda bayes_params: mod_log_likelihood(bayes_params, map_params))(samples['params']))

    value_and_grad_fn = value_and_grad(loss)
    for _ in range(100):
        value, g = value_and_grad_fn(map_params)
        map_params = map_params - g
        print(value, jnp.linalg.norm(g))
    return map_params

```
